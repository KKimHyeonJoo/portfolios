```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
movies = pd.read_csv('movies.csv')
movies.head(3)
```

```python
movies.shape
```

```python
movies.info()
```

```python
movies.hist(bins=50, figsize=(20,15)) # 히스토그램
plt.show()
```

# 1) score 정수형 변환

```python

movies.score.value_counts()
```

```python
movies['score'].replace('Hit Movie', 0, inplace=True)
movies['score'].replace('Average Movie', 1, inplace=True)
movies['score'].replace('Flop Movie', 2, inplace=True)
movies
```

```python
movies.isnull().sum()
```

# 2) 결측값

```python
movies.replace({r'\N': np.nan}, inplace=True)
movies.isnull().sum()
```

### - startYear, runtimeMinutes는 평균값으로 대체

```python
movies['startYear'] = pd.to_numeric(movies['startYear'], errors='coerce')  
# errors='coerce'는 숫자로 변환할 수 없는 값을 NaN으로 처리
movies['startYear'] = movies['startYear'].fillna(movies['startYear'].mean())
```

```python
movies['runtimeMinutes'] = pd.to_numeric(movies['runtimeMinutes'], errors='coerce')  
# errors='coerce'는 숫자로 변환할 수 없는 값을 NaN으로 처리
movies['runtimeMinutes'] = movies['runtimeMinutes'].fillna(movies['startYear'].mean())
```

```python
movies.isnull().sum()
```

### - 결측값 있는 행 삭제

```python
movies = movies.dropna(axis=0)
movies.isnull().sum()
```

```python
movies.info()
```

```python
movies
```

# 3) 정수로 되어있는 문자열을 정수형으로 변환

```python
movies['startYear'] = movies['startYear'].astype(int)
movies['runtimeMinutes'] = movies['runtimeMinutes'].astype(int)

movies['first_director'] = movies['first_director'].astype(int)
movies['first_writer'] = movies['first_writer'].astype(int)

movies['first_actor'] = movies['first_actor'].astype(int)
movies['first_actress'] = movies['first_actress'].astype(int)

movies.dtypes
```

```python
plt.figure(figsize=(14,10))
sns.boxplot(x='averageRating',y='genres_first',data=movies)
```

```python
# genres_first 열의 고유한 값들을 가져옵니다.
unique_genres = movies['genres_first'].unique()

# 각 장르에 대한 정수 레이블을 만듭니다.
genre_to_int = {genre: i for i, genre in enumerate(unique_genres)}

# genres_first 열을 정수로 매핑합니다.
movies['genres_first_encoded'] = movies['genres_first'].map(genre_to_int)
movies.head(2)
```

## 5) titleType 정수형 변환

```python
movies.titleType.value_counts()
```

```python
movies['titleType'].replace('movie', 0, inplace=True)
movies['titleType'].replace('tvMovie', 1, inplace=True)
movies.info()
```

# 6) 모델에 필요없는 열 삭제

```python
movies = movies.drop(['genres', 'genres_first'], axis=1)
movies.reset_index(drop=True, inplace=True)
movies.head()
```

```python
movies.info()
```

```python
movies.describe()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

movies.hist(bins=50, figsize=(20,15)) # 히스토그램
plt.show()
```

```python
corr = movies.corr()
mask = np.zeros(corr.shape, dtype=bool)
mask[np.triu_indices(len(mask))] = True
plt.subplots(figsize=(20,15))
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns,cmap='RdYlGn',annot=True,mask = mask)
```
